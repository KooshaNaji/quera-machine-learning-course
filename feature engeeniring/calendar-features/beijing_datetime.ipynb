<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
ابتدا نیاز است فایل مجموعه‌داده‌ی بدون مقادیر گم‌شده (<code>beijing_no_missing.csv</code>) را بخوانیم.
</p>

```python
import pandas as pd
# reading the dataset
df = pd.read_csv('beijing_no_missing.csv', encoding='gbk', low_memory=False, parse_dates=['tradeTime'])
```

```python
df
```

<h2 dir="rtl" align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
استخراج عدد روز، ماه و سال
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
در سلول زیر، روز، ماه و سال انجام معامله را به ترتیب با اسم‌های <code>tradeDay</code> و <code>tradeMonth</code> و <code>tradeYear</code> به مجموعه‌داده اضافه می‌کنیم:
</font>
</p>

```python
df['tradeDay'] = df['tradeTime'].dt.day
df['tradeMonth'] = df['tradeTime'].dt.month
df['tradeYear'] = df['tradeTime'].dt.year

df[['tradeTime', 'tradeDay', 'tradeMonth', 'tradeYear']].head()
```

<h2 dir="rtl" align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
استخراج نام روز
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
در سلول زیر اسم روز (شنبه، یک‌شنبه و …) را به عنوان یک ویژگی دیگر با اسم <code>tradeDayName</code> به مجموعه‌داده اضافه می‌کنیم:

</font>
</p>

```python
df['tradeDayName'] = df['tradeTime'].dt.day_name()

df[['tradeTime', 'tradeDayName']].head()
```

<h2 dir="rtl" align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
استخراج آخر سال بودن یا نبودن
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
به کمک سلول زیر، می‌توانیم آخر سال بودن یا نبودن یک تاریخ را به عنوان یک ویژگی جدید، با اسم <code>isTradeEndYear</code> اضافه کنیم:

</font>
</p>

```python
df['isTradeEndYear'] = df['tradeTime'].dt.is_year_end

df[['tradeTime', 'isTradeEndYear']].head()
```


<h2 dir="rtl" align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
کشف فصل (گرم یا سرد)
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
در سلول زیر کدی می‌نویسیم که مشخص می‌کند <code>tradeTime</code> در فصل‌های گرم سال اتفاق افتاده است یا خیر. این ویژگی را با اسم <code>isHotMonth</code> اضافه می‌کنیم:

</font>
</p>

```python
hot_weather = ['March', 'April', 'May', 'June', 'July', 'August']
cold_weather = ['September', 'October', 'November', 'December', 'January', 'February']

df['isHotMonth'] = [month in hot_weather for month in df['tradeTime'].dt.month_name()]

df[['tradeTime', 'isHotMonth']].head()
```


<h2 dir="rtl" align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
تقسیم‌بندی براساس سال ساخت
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
می‌خواهیم ساختمان‌ها را بر اساس سال ساخت‌شان به سه دسته تقسیم کنیم. دسته اول را به ساختمان‌هایی اختصاص می‌دهیم که بین سال‌های ۱۹۰۶ تا ۱۹۸۰ ساخته شده باشند. دسته دوم مخصوص ساختمان‌هایی است که سال ساختشان بین ۱۹۸۱ تا ۲۰۱۰ باشد و در دسته سوم، ساختمان‌هایی قرار می‌گیرند که [تقریباً] نوساز به حساب می‌آیند. سال ساخت این ساختمان‌ها باید حداقل ۲۰۱۱ باشد. این سه دسته را به ترتیب به سه عدد ۱، ۲ و ۳ نسبت می‌دهیم. در سلول زیر، ابتدا جنس ستون <code>constructionTime</code> را به <code>int</code> تغییر می‌دهیم:
</font>
</p>

```python
df['constructionTime'] = df['constructionTime'].astype('int')

def categorize_year(year):
    if year <= 1980:
        return 1
    if year <= 2010:
        return 2
    return 3

df['buildingLife'] = df['constructionTime'].apply(categorize_year)

df[['constructionTime', 'buildingLife']].head()
```


<h2 dir="rtl" align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
استخراج تعطیلی روز معامله
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
ویژگی دیگری که می‌توان استخراج کرد، تعطیل بودن یا نبودن روزی است که در آن معامله انجام شده است. به این منظور می‌توانیم از کتابخانه‌ی <code>holidays</code> استفاده کنیم.
</font>
</p>

```python
%pip install holidays
import holidays
china_holidays = holidays.China()
df['isDayOff'] = df['tradeTime'].apply(lambda date : date in china_holidays)

df[['tradeTime', 'isDayOff']].head()
```
